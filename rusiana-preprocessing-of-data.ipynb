```python
#load the datasets
df = pd.read_csv('/kaggle/input/gender-classification-mg/gender_classification_v7.csv')
```

```python
#drop if there are categorical values aside from target class
x = df[['long_hair', 'forehead_width_cm', 'forehead_height_cm', 'nose_wide', 'nose_long', 'lips_thin', 'distance_nose_to_lip_long']]
y = df['gender']
```

```python
#x and y split into training and test each
from sklearn.model_selection import train_test_split

#splitting into 85% train & 15% test
X_train, X_test,  y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=123 )
```

```python
#shape of x train & test and y train & test
print("The shape of x train: ", X_train.shape)
print("The shape of x test: ", X_test.shape)
print("The shape of y train: ", y_train.shape)
print("The shape of y test: ", y_test.shape)
```

```python
#Perform Standardization

numerical_features = ['long_hair', 'forehead_width_cm', 'forehead_height_cm', 'nose_wide', 'nose_long', 'lips_thin', 'distance_nose_to_lip_long']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Use the fit method to perform the transformation
scaler.fit(X_train [numerical_features])

# Use the transform method to perform the transformation
X_train.loc[:, numerical_features] = scaler.transform(X_train [numerical_features])

```

```python
#Result of standardization rounded to 4 decimal places
X_train.describe().round(4)
```
